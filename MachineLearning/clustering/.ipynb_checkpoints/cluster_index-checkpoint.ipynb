```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score,davies_bouldin_score
```

```python
iris_data=pd.read_csv("Iris.csv")
```

```python
features=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]
```

```python
iris_data=iris_data.dropna(subset=features)
```

```python
train_data=iris_data[features]
```

```python
mean=train_data.mean()
std=train_data.std()
```

```python
train_data=(train_data-mean)/std
```

```python
wcss=[]
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(train_data) 
    wcss.append(kmeans.inertia_)

```

```python
plt.plot(range(1,11),wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
```

```python
n_cluster=4 ## As we have elbow at 4
```

```python
kmeans = KMeans(n_clusters = n_cluster, init = 'k-means++',random_state=45)
kmeans.fit(train_data) 
```

```python
labels=kmeans.predict(train_data)
```

```python
kmeans.cluster_centers_
```

```python
clusters = {}
n = 0
for item in labels:
    if item in clusters:
        clusters[item].append(n)
    else:
        clusters[item] = [n]
    n +=1
```

```python
for cluster_number,index in clusters.items():
    print("Total Number of data points in cluster",cluster_number,"is:-",len(clusters[cluster_number]))
```

```python
Silhouettescore=silhouette_score(train_data,labels)
davisbouldin=davies_bouldin_score(train_data,labels)

print("silhouette_score is :-",Silhouettescore)
print("davies_bouldin  index is :-",davisbouldin)
```
