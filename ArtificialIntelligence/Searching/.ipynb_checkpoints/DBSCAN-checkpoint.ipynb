```python
def NumberOfOutliers(columns):
    count=0
    lower_range=columns.mean()-columns.std()
    upper_range=columns.mean()+columns.std()
    
    count+=(columns>=upper_range).sum()
    count+=(columns<=lower_range).sum()
    
    return (count/columns.count())*100
    
```

```python
outlier_count=student_data.apply(lambda x:NumberOfOutliers(x))
```

```python
outlier_count.sort_values()
```

```python
def dbscanClustering(data,eps,min_samples):
    clustering=DBSCAN(eps=eps,min_samples=min_samples)
    result=clustering.fit(student_data)
    labels=result.labels_
    score=silhouette_score(data, labels)
    print("NUmber of cluster is :-",len(set(labels)))
    print("silhouette score is:-",score)
    return [labels,score]
    
```

```python
def hierarchicalClustering(data,ClusterNumber,linkage,distance_threshold):
    clusters=AgglomerativeClustering(n_clusters=None,linkage=linkage,distance_threshold=distance_threshold)
    result=clusters.fit(data)
    lables=result.labels_
    score=silhouette_score(data,lables)
    print("Number of cluster is :-",len(set(lables)))
    print("silhouette score is:-",score)
    return [lables,score,result]
    
    
    
    
```
